```python

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py
```

```python
df = pd.read_csv('./HOTSTAR.csv')
```

```python
df=df[df['title'].notna()]
```

```python
for i in df.columns:
    null_rate = df[i].isna().sum() / len(df) * 100 
    if null_rate > 0 :
        print("{} null rate: {}%".format(i,round(null_rate,2)))
```

```python
df= df.drop(['metascore'], axis=1)
```

```python
df= df.drop(['awards'], axis=1)
```

```python
print(' \n Venues Played At:',df['type'].unique())     
```

```python
dft=df[df['imdb_rating'].notna()]

features = ['imdb_rating']
dft[features].hist( color = 'blue', edgecolor = 'black',cumulative=true);
```

```python
dft=df[df['imdb_rating'].notna()]

features = ['imdb_rating']
dft[features].hist( color = 'blue', edgecolor = 'black');
```

```python
a =  df.sort_values('imdb_rating',ascending=False).head(30)
typ = list(a['type'])
courses = list(a['title'])
values = list(a['imdb_rating'])
col = []
for val in typ :
    if val == 'movie':
        col.append('blue')
    elif val =='series':
        col.append('green')
    else:
        col.append('red')

plt.figure(figsize=(12, 8))
ax = a['imdb_rating'].plot(kind='bar',color=col)
ax.set_title('IMDB RATING')
ax.set_xlabel('Movies/ Tv Shows')
ax.set_ylabel('IMDB RATING')
ax.set_xticklabels(courses)


for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.15, p.get_height()+0.25))
    
```

```python
dft=df[df['country'].notna()]
dft.country = dft.country.str.replace(' ', '')
dft=dft[df['title']=='movie'].country.str.get_dummies(',').stack().sum(level=1).sort_values(ascending=False).head(10)
ax=dft.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x(), p.get_height()+0.25))
```

```python
dft=df[df['rated'].notna()]
dft.rated = dft.rated.str.replace(' ', '')
dft=dft.rated.str.get_dummies(',').stack().sum(level=1).sort_values(ascending=False).head(10)
ax=dft.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x(), p.get_height()+0.25))
```

```python
dft=df[df['director'].notna()]
dft.director = dft.director.str.replace(' ', '')
dft=dft.director.str.get_dummies(',').stack().sum(level=1).sort_values(ascending=False).head(10)
ax=dft.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x(), p.get_height()+0.25))
```

```python
dft=df[df['year'].notna()]
dft.year = dft.year.str.replace(' ', '')
dft=dft.year.str.get_dummies(',').stack().sum(level=1).sort_values(ascending=False).head(10)
ax=dft.plot(kind='bar')
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x(), p.get_height()+0.25))
```
